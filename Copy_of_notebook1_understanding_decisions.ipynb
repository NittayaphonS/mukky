**Code your solution here:**

````python
# Your study recommendation system

# Step 1: Collect input from the user
hours_studied = float(input("How many hours have you studied today? "))
stress_level = input("What is your stress level (Low, Medium, High)? ").capitalize()

# Step 2: Determine recommendation based on hours studied and stress level
if hours_studied < 2 and stress_level == "Low":
    recommendation = "Focus on studying more today"
elif hours_studied >= 2 and stress_level == "High":
    recommendation = "Take a short break and relax"
elif hours_studied >= 2 and stress_level == "Low":
    recommendation = "Keep going, maintain your pace"
else:
    recommendation = "Maintain balance and adjust your study plan"

# Step 3: Display the recommendation
print("\nRecommendation:", recommendation)
````

````python
hours_studied = float(input("How many hours have you studied today? "))
stress_numeric = int(input("On a scale of 1-10, how stressed are you? "))

# Map numeric stress to Low, Medium, High
if 1 <= stress_numeric <= 3:
    stress_level = "Low"
elif 4 <= stress_numeric <= 6:
    stress_level = "Medium"
else:
    stress_level = "High"

# Decision logic
if hours_studied < 2 and stress_level == "Low":
    recommendation = "Focus on studying more today"
elif hours_studied >= 2 and stress_level == "High":
    recommendation = "Take a short break and relax"
elif hours_studied >= 2 and stress_level == "Low":
    recommendation = "Keep going, maintain your pace"
else:
    recommendation = "Maintain balance and adjust your study plan"

# Output
print("\nRecommendation:", recommendation)
````

## Activity 3: Debugging and Refinement (10 minutes)

### The Broken Code Challenge

Here's some code that doesn't work as intended. Your job: **Figure out what's wrong by thinking through the logic first**.

````python
student_gpa = 3.2

if student_gpa >= 2.0:
    status = "Probation"
elif student_gpa >= 3.0:
    status = "Good Standing"
elif student_gpa >= 3.5:
    status = "Dean's List"
else:
    status = "Failing"

print(f"Status: {status}")
````

**Step 1: Trace Through the Logic**

What happens when `student_gpa = 3.2`?

- Check 1: `3.2 >= 2.0`? → (double-click to write your answer)
- If true, what happens next? →

**Step 2: Identify the Problem**

What's wrong with this logic? Write your analysis (double-click to edit):

The problem is: The conditions are checked in the wrong order. The first condition if student_gpa >= 2.0: is too broad, so it captures all GPAs above 2.0 and prevents the later elif statements from running.

It happens because: Python evaluates if/elif/else in order, and the first true condition stops the rest.

**Step 3: Fix It**

How would you fix this code? Write your corrected version:

````python
# Your fixed version here
student_gpa = 3.2

if student_gpa >= 3.5:
    status = "Dean's List"
elif student_gpa >= 3.0:
    status = "Good Standing"
elif student_gpa >= 2.0:
    status = "Probation"
else:
    status = "Failing"

print(f"Status: {status}")
````
