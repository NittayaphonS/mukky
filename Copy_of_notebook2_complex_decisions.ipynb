### Your Mission: Build a College Application Decision System

Universities use complex criteria to make admission decisions. Your job is to design and implement a fair, logical system.

**Step 1: Research and Design** (8 minutes)

Think about what factors matter for college admissions:

- Academic performance (GPA, test scores)
- Extracurricular activities

**Design your system:**

> **Write your admission criteria (double-click to edit):**

```
Factor 1: GPA (Academic performance)

Factor 2: Extracurricular activities (number of activities)

Factor 3: Personal essay score (1–10)

Decision logic: A student is Admitted if at least 2 out of 3 factors meet the threshold:
GPA ≥ 3.5 → Pass
Extracurricular activities ≥ 2 → Pass
Essay score ≥ 7 → Pass
Otherwise, student is Waitlisted
If none of the criteria are met, student is Rejected

If exactly 2 factors are met → Admitted
If only 1 factor is met → Waitlisted
All below thresholds → Rejected
```

Now code your admissions system. You can:

**Your implementation:**

````python
# College Admissions Decision System
# Your code here
# College Admissions Decision System

# Step 1: Collect input
gpa = float(input("Enter GPA (0.0 - 4.0): "))
extracurriculars = int(input("Number of extracurricular activities: "))
essay_score = int(input("Essay score (1-10): "))

# Step 2: Evaluate each criterion
criteria_met = 0

if gpa >= 3.5:
    criteria_met += 1
if extracurriculars >= 2:
    criteria_met += 1
if essay_score >= 7:
    criteria_met += 1

# Step 3: Determine admission decision
if criteria_met == 3 or criteria_met == 2:
    decision = "Admitted"
elif criteria_met == 1:
    decision = "Waitlisted"
else:
    decision = "Rejected"

# Step 4: Display the result
print(f"\nAdmission Decision: {decision}")
````

## Activity 3: Debugging Complex Logic (10 minutes)

### The Mystery: Why Isn't This Working?

A student created this scholarship eligibility system, but it's not working as expected:

````python
print("=== Scholarship Eligibility Checker ===")

gpa = float(input("Enter GPA (0.0-4.0): "))
community_hours = int(input("Community service hours: "))
financial_need = input("Have financial need? (yes/no): ").lower()
essay_score = int(input("Essay score (1-10): "))

# Should be eligible if:
# - GPA 3.0+ AND (community hours 50+ OR financial need) AND essay 7+

if gpa >= 3.0 and community_hours >= 50 or financial_need == "yes" and essay_score >= 7:
    print("Eligible for scholarship!")
else:
    print("Not eligible")
````

**Your debugging process:**

**Step 1: Understand the Intent**

What is this code supposed to do? Write the logic in plain English (double-click to edit): GPA ≥ 3.0 AND (community_hours ≥ 50 OR financial_need) AND essay_score ≥ 7

Should be eligible if:

- The student’s GPA is 3.0 or higher, AND

- The student has 50 or more community service hours OR the student has financial need, AND
- The student’s essay score is 7 or higher

**Step 2: Trace Through the Logic**

What does this condition actually check?

gpa >= 3.0 and community_hours >= 50 or financial_need == "yes" and essay_score >= 7

```

Think about operator precedence. What gets evaluated first?

- and is evaluated before or

**Step 3: Identify the Problem (double-click to edit):**

The problem is: The or and and operators are combined without parentheses, so Python evaluates and first, breaking the intended grouping.

The fix is: Use parentheses to group the OR condition correctly:

**Step 4: Write the Corrected Version**

````python
# Your corrected code here
print("=== Scholarship Eligibility Checker ===")

# Inputs
gpa = float(input("Enter GPA (0.0-4.0): "))
community_hours = int(input("Community service hours: "))
financial_need = input("Have financial need? (yes/no): ").lower()
essay_score = int(input("Essay score (1-10): "))

# Corrected eligibility condition
if gpa >= 3.0 and (community_hours >= 50 or financial_need == "yes") and essay_score >= 7:
    print("Eligible for scholarship!")
else:
    print("Not eligible")
````

Operator Precedence in Python

Python evaluates logical operators in a specific order:

- not – highest priority
- and – evaluated next
- or – lowest priority

This means that and operations are computed before or unless you use parentheses to change the order.
