## Part 2: Financial Decision Logic (15 minutes)

### Budget Limit Checker
Apply your if/else skills to real budget management:

````python
print("=== Budget Limit Checker ===")

# Get user's budget and expense
monthly_budget = float(input("Enter your monthly budget: $"))
expense = float(input("Enter expense amount: $"))

# Decision logic here - you write this part!
# If expense > 50% of budget: "Major purchase - think carefully"
if expense > 0.5 * monthly_budget:
    print("Major purchase - think carefully")
# If expense > 25% of budget: "Significant expense"
elif expense > 0.25 * monthly_budget:
    print("Significant expense")
# If expense > 10% of budget: "Moderate expense"
elif expense > 0.1 * monthly_budget:
    print("Moderate expense")
# Otherwise: "Minor expense - within budget"
else:
    print("Minor expense - within budget")

print("Budget check complete.")
````

````python
print("=== Savings Goal Tracker ===")

goal_amount = float(input("Savings goal: $"))
current_savings = float(input("Current savings: $"))

# Your task: Calculate percentage and give encouraging feedback
percentage = (current_savings / goal_amount) * 100

# 100%+: "Goal achieved! Consider a new goal."
if percentage >= 100:
    print("Goal achieved! Consider a new goal.")
# 75-99%: "Almost there! You're doing great."
elif percentage >= 75:
    print("Almost there! You're doing great.")
# 50-74%: "Halfway there - keep it up!"
elif percentage >= 50:
    print("Halfway there - keep it up!")
# 25-49%: "Good progress, stay consistent"
elif percentage >= 25:
    print("Good progress, stay consistent")
# Under 25%: "Just getting started - you've got this!"
else:
    print("Just getting started - you've got this!")

print(f"You've reached {percentage:.2f}% of your goal.")
````

## Part 3: Smart Purchase Advisor (20 minutes)

### The Challenge
Build a program that helps users make smart purchase decisions by considering multiple factors.

### Requirements

Your purchase advisor should:

1. Get purchase amount and user's available money
2. Check if they can afford it (basic affordability)
3. Give different advice based on purchase size
4. Consider their financial situation

### Starter Framework

````python
print("=== Smart Purchase Advisor ===")

# Get financial information
available_money = float(input("How much money do you have available? $"))
purchase_price = float(input("How much does the item cost? $"))
monthly_income = float(input("What's your monthly income? $"))

# Calculate affordability percentage
affordability_ratio = purchase_price / available_money

# Your task: Build the decision logic
# Consider:
# - Can they afford it at all?
# - Is it a large percentage of their available money?
# - Is it reasonable compared to their income?
# - What advice should you give?

# Suggestion: Use multiple if/elif conditions to give personalised advice
````

````python
print("=== Smart Purchase Advisor ===")

# Get financial information
available_money = float(input("How much money do you have available? $"))
purchase_price = float(input("How much does the item cost? $"))
monthly_income = float(input("What's your monthly income? $"))

# Bonus: category
category = input("Is this a 'need' or a 'want'? ").lower()

# Bonus: emergency fund
emergency_fund = float(input("How much is your emergency fund? $"))

if available_money == 0 or monthly_income == 0:
    print("Invalid financial data. Please check your inputs.")
else:
    affordability_ratio = purchase_price / available_money

    # 1. Can't afford
    if purchase_price > available_money:
        needed = purchase_price - available_money
        print(f" Save up first - you need ${needed:.2f} more")

    else:
        # 2. Can afford but expensive
        if affordability_ratio > 0.5:
            print(f"⚠️ You can afford it, but it's {affordability_ratio*100:.2f}% of your available money.")
            print("Consider waiting.")

        # 3. Reasonable purchase
        elif affordability_ratio > 0.2:
            print("💡 This seems like a reasonable purchase for your budget")

        # 4. Small purchase
        else:
            print("✅ This is a minor expense - go for it!")

        # 🔒 Bonus: Emergency fund check
        remaining_money = available_money - purchase_price
        if remaining_money < emergency_fund:
            print("🚨 Warning: This purchase may reduce your emergency savings!")

        # 🛒 Bonus: Category-based advice
        if category == "want" and affordability_ratio > 0.3:
            print("This is a luxury item - consider if it's really necessary.")
        elif category == "need":
            print("This is a necessary expense.")

        # 💳 Bonus: Payment plan (for expensive items)
        if purchase_price > 0.3 * monthly_income:
            months = 6
            monthly_payment = purchase_price / months
            print(f"💳 Payment plan: ${monthly_payment:.2f}/month for {months} months")

        # Extra info
        print(f"After purchase, you'll have ${remaining_money:.2f} left.")
````

import pandas as pd

# ----------------------
# 1. Create sample data (or replace with your CSV)
# ----------------------
data = {
    "Description": ["McDonalds", "Uber ride", "Netflix"],
    "Amount": [20, 15, 10]
}

df = pd.DataFrame(data)

# ----------------------
# 2. Categorise expenses
# ----------------------
def categorise(desc):
    desc = desc.lower()
    if "uber" in desc:
        return "Transport"
    elif "netflix" in desc:
        return "Entertainment"
    else:
        return "Food"

df["Category"] = df["Description"].apply(categorise)

# ----------------------
# 3. Set budget
# ----------------------
budget = {
    "Food": 50,
    "Transport": 50,
    "Entertainment": 20
}

# ----------------------
# 4. Calculate spending
# ----------------------
spending = df.groupby("Category")["Amount"].sum()

# ----------------------
# 5. Compare with budget + feedback
# ----------------------
print("📊 Result:\n")

for cat in budget:
    used = spending.get(cat, 0)
    percent = (used / budget[cat]) * 100

    if percent > 100:
        status = "⚠️ Over budget"
    elif percent > 80:
        status = "⚡ Near limit"
    else:
        status = "✅ OK"

    print(f"{cat}: ${used} ({percent:.1f}%) -> {status}")

# ----------------------
# 6. Show final table
# ----------------------
print("\n📋 Data:")
print(df)
